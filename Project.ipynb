```python
import pandas as pd
#read csv
df = pd.read_csv('leetcode_dataset - lc.csv')
```

organize the dataset based on acceptance rate

```python
df = df.sort_values(by=['acceptance_rate'], ascending=False)
```

Convert the company column to a numeric value

```python
company_dict = {'Google': 1, 'Amazon': 2, 'Microsoft': 3, 'Facebook': 4}
df['company_num'] = df['companies'].map(company_dict)

from sklearn.feature_extraction.text import TfidfVectorizer
```

Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'

```python
tfidf = TfidfVectorizer(stop_words='english')
```

Construct the required TF-IDF matrix by fitting and transforming the data

```python
df['overview'] = df['description'].fillna('')
tfidf_matrix = tfidf.fit_transform(df['overview'])

from sklearn.metrics.pairwise import linear_kernel
```

Compute the cosine similarity matrix

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
indices = pd.Series(df.index, index=df['title']).drop_duplicates()

def get_recommendations(title, cosine_sim=cosine_sim):
# Get the index of the LeetCode problem that matches the title
    idx = indices[title]
    # Get the pairwsie similarity scores of all LeetCode problems with that description
    sim_scores = list(enumerate(cosine_sim[idx]))

# Sort the problems based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

# Get the scores of the 10 most similar problems
    sim_scores = sim_scores[1:11]

# Get the problem indices
    title_indices = [i[0] for i in sim_scores]

# Return the top 10 most similar problems
    return df['title'].iloc[title_indices]
```

Combine the description and company columns into a single overview column

```python
df['overview'] = df['description'].fillna('') + ' ' + df['company_num'].fillna('').astype(str)
```

Construct the required TF-IDF matrix by fitting and transforming the data

```python
tfidf_matrix = tfidf.fit_transform(df['overview'])
```

Split the dataset into training and testing sets

```python
#define train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, df['acceptance_rate'], test_size=0.2, random_state=42)

from sklearn.metrics.pairwise import linear_kernel
```

Train the model on the training set

```python
cosine_sim = linear_kernel(X_train, X_train)
```

Compute the similarity scores on the testing set

```python
sim_scores = linear_kernel(X_test, X_train)
```

Calculate the MSE and RMSE

```python
#import mean_squared_error and math
from sklearn.metrics import mean_squared_error
import math
mse = mean_squared_error(y_test, sim_scores.dot(y_train))
rmse = math.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
```

## Final version of code
The code below is the final verion of the program. Everything before was used to understand, manipulate, and test the functonality of the code below. The code can be extracted from the cell below and ran independently as long as the csv file is present in the same directory as the code.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# read data
df = pd.read_csv('leetcode_dataset - lc.csv')
#drop solution_link column, related_topics column, similar_questions column
df = df.drop(['solution_link', 'related_topics', 'similar_questions'], axis=1)
# convert company column to string
df['description'] = df['description'].astype(str)
df['companies'] = df['companies'].astype(str)

# vectorize the features
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df[['description', 'companies', 'difficulty']].apply(lambda x: ' '.join(x), axis=1))

# get user preferences
company_pref = input("Enter company preference: ")
desc_pref = input("Enter description preference: ")
diff_pref = input("Enter difficulty preference (Easy, Medium, Hard): ")

# transform user preferences
user_pref = [desc_pref, company_pref, diff_pref]
user_pref = vectorizer.transform([' '.join(user_pref)])

# calculate similarity scores
cosine_sim = cosine_similarity(X, user_pref)

# get indices of top 10 most similar problems
sim_scores = list(enumerate(cosine_sim))
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[:10]
problem_indices = [i[0] for i in sim_scores]

# print recommended problems
print("Recommended problems:")
for idx in problem_indices:
    print(df.iloc[idx]['title'])
```

#### Data Analysis

```python
import numpy as np
import pandas as pd

# Read the data
data = pd.read_csv('leetcode_dataset - lc.csv')
data.head()
```

```python
# Get the number of Easy, Medium, and Hard problems
data['difficulty'].value_counts()
```

```python
# get the names of all of the companies (the companies column contains a list of companies) and return a list of unique companies in alphabetical order
companies = data['companies'].str.split(',').explode().unique().tolist()
companies
```

```python
# get the number of problems for each company
num_problems_per_company = data['companies'].str.split(',').explode().value_counts()
num_problems_per_company
```

```python
# get the number of problems that don't have a company associated with them
num_problems_no_company = data['companies'].isna().sum()
num_problems_no_company
```

```python
# get the number of empty values in each column
data.isna().sum()
```
